## üé≤ One Betting Strategy Versus All Others

This program computes the winning probabilities **exactly**,
$$ P_A(S,T),\ P_B(S,T),\ P_U(S,T) $$  
where $S$ is fixed and $T$ ranges over all possible betting strategies with the same total number of chips as $S$.

- The strategy $S$ is selected. Then all possible betting strategies with the same total number of chips are generated.
- All stacked bar charts for $P_A(S,T)$ and $P_B(S,T)$ are displayed.
- The computed values are saved to a CSV file.
- The plot is saved as a (editable) PDF graphic.
- For more than 28 betting strategies, the x-axis labels overlap. In that case, it is recommended to hide the plot by setting `plot=False`.

---

### ‚úÖ Requirements

- Exactly **one die** with user-defined outcome probabilities is used; the probabilities sum to 1.

---

### ‚öôÔ∏è Inputs

- `p`: probabilities of the individual fields  
- `S`: base betting strategy

```python
from fractions import Fraction
import chip_strategies as ps
# ========= Eingaben =================

"""
Visualizes stacked bar charts for P_A(S,T) and P_B(S,T)
for a fixed strategy S against all opponent strategies T.

Args:
    S (tuple[int]): Fixed strategy of player A.
    p_vals (list[float] or list[Fraction]): Outcome probabilities p_i.
    xlabel, ylabel, title (str): Axis labels and plot title.
    width (float): Bar width (recommended: 0.6 to 1.0).
    sort_by (str): 'PA' or 'PB'; sort by descending values.
    save_pdf (str): Filename to save the plot as a PDF.
    csv_path (str): Filename to save results as a CSV file.
"""

# ============================================================
# Input
# ============================================================

# Input of outcome probabilities
p = (1/2,1/3,1/6)

# Input of the base betting strategy S,
# which is played against all other strategies

S = (3,2,1)

# ============================================================
# Consistency check
# ============================================================
# exact=True: use exact rational arithmetic
ps.p_list = ps.prepare_p_list(p, exact=True)
assert len(ps.p_list) == len(S), "p_list und V m√ºssen dieselbe L√§nge haben!"

# ============================================================
# Visualization
# ============================================================

# Sorted by PB; saves PDF and CSV output.
# For more than 28 strategies, plot=False is recommended
# since x-axis labels will overlap.

ps.P_fixiert(
    S, p,
    xlabel="",
    ylabel=fr"$P_A(S,T), \;\; P_B(S,T)$",
    title=fr"Stacked $P_A\; / \;P_B$ for S={S}",
    width=0.3,
    sort_by='PA',
    save_pdf="stacked_PA_PB.pdf",
    breite=7,
    hoehe=3.5,
    csv_path="results_PA_PB.csv",
    plot=True    # plot=False: no plot
)
```
